```python
from langchain_community.document_loaders import PyPDFLoader , TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_loader = TextLoader('speech.txt')
text_doc = text_loader.load()
```

```python
from IPython.display import Markdown

Markdown(text_doc[0].page_content)
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500 , chunk_overlap=100
)

chunk = splitter.split_documents(text_doc)
chunk
```

```python
Markdown(chunk[0].page_content)
```

```python

embeddings.embed_query(chunk[0].page_content)
```

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(
    chunk,
    embeddings
)

db
```

```python
query="It will be all the easier for us to conduct ourselves as belligerents"
retrieved_result = db.similarity_search(query)

retrieved_result
```
