```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter
import os 
from dotenv import load_dotenv
load_dotenv

os.environ['HF_TOKEN'] = os.getenv('HF_TOKEN')
os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
```

```python

class DataIngestion:
    def __init__(self , pdf_path = "C:\Users\user\Desktop\Gen_AI\RAG\The_GALE_ENCYCLOPEDIA_of_MEDICINE_SECOND.pdf" , chunk_size = 1000 ,chunk_overlap = 500):
        self.pdf_path = pdf_path
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.data_loading = PyPDFLoader(self.pdf_path)
        self.text_splitter = RecursiveCharacterTextSplitter(chunk_size=self.chunk_size , chunk_overlap=self.chunk_overlap)
        self.embedding = OpenAIEmbeddings() 

    def process_documents(self):
        docs = self.data_loading.load()[:50]
        return self.text_splitter.split_documents(docs)
    

class Retrieval(DataIngestion):
    def __init__(self):
        super().__init__()

    def retrieval(self):
        pdf_chunk = self.process_documents()
        vectors = FAISS.from_documents(pdf_chunk , self.embedding)
        return vectors
    



    
    
```

```python
loader = DataIngestion()


class VectorStore:
    def __init__(self):
        self.loader = loader.process_documents()
        self.embedding = loader.embedding
        self.vectordb = FAISS.from_documents(self.loader , self.embedding)

    
    def get_retriever(self):
        return self.vectordb.as_retriever(search_kwargs={"k":5})
    
    def similarity_search(self , query):
        return self.vectordb.similarity_search(query , k=5)
    
    def similarity_search_with_score(self , query):
        return self.vectordb.similarity_search_with_score(query , k=5)
    
    def is_initialized(self):
        return self.vectordb is not None
```

```python
vectordb = VectorStore()

similarity_search = vectordb.similarity_search("What is generative AI?")

similarity_search
```

```python
import logging
import time
import openai  # or any GenAI SDK
from dotenv import load_dotenv
import os 

load_dotenv()
from langchain_openai import ChatOpenAI

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(name)s - %(message)s',
    handlers=[
        logging.FileHandler("genai_app.log"),
        logging.StreamHandler()
    ]
)

logger = logging.getLogger("GenAIApp")

# Simulated GenAI function
def generate_response(prompt: str) -> str:
    try:
        logger.info(f"Received prompt: {prompt}")
        
        start_time = time.time()
        
        # Simulate API call (e.g., OpenAI)
        model = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0.7)
        response = model.invoke(prompt)
        # response = openai.Completion.create(...)
        # For demo:
        response_text = f"Generated response to: {prompt[:50]} is  : {response.content}"
        
        latency = time.time() - start_time
        logger.info(f"Response generated in {latency:.2f}s")
        logger.debug(f"Full response: {response_text}")
        
        return response_text

    except Exception as e:
        logger.error(f"Error generating response for prompt: {prompt}", exc_info=True)
        return "Sorry, I couldn't process your request."

# Example usage
if __name__ == "__main__":
    user_prompt = "Explain quantum computing in simple terms."
    response = generate_response(user_prompt)
    logger.info(f"Final response sent to user: {response}")
```
